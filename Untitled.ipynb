```python
import duckdb
import pandas as pd
import os

# Path setup
data_path = "../data/"
parquet_files = [
    "train_data.parquet",
    "test_data.parquet",
    "add_event.parquet",
    "add_trans.parquet",
    "offer_metadata.parquet"
]

# Step 1: Load the dictionary and create mapping
dict_df = pd.read_csv(os.path.join(data_path, "data_dictionary.csv"))
rename_dict = dict(zip(dict_df['masked_column'], dict_df['Description']))

# Step 2: Initialize DuckDB in-memory connection
con = duckdb.connect(database=':memory:')

# Step 3: Read and rename columns dynamically
dfs = {}

for file in parquet_files:
    table_name = file.replace(".parquet", "")
    file_path = os.path.join(data_path, file)
    
    # Read into DataFrame
    df = con.execute(f"SELECT * FROM parquet_scan('{file_path}')").df()
    
    # Rename using mapping ONLY for columns that exist
    df.rename(columns={col: rename_dict[col] for col in df.columns if col in rename_dict}, inplace=True)
    
    # Store
    dfs[table_name] = df
    
    # Quick EDA summary
    print(f"\n=== {table_name} ===")
    # print("Shape:", df.shape)
    # print("Columns:", df.columns.tolist())
    # print(df.head(2))
    temp_view = f"{table_name}_df"
    con.register(temp_view, df)
    
    # Create permanent DuckDB table from the temp view
    con.execute(f"CREATE OR REPLACE TABLE {table_name} AS SELECT * FROM {temp_view}")
    
    print(f"✅ Table '{table_name}' created with shape {df.shape}")
```

```python
# Get number of clicks per customer per offer
clicks = con.execute("""
    SELECT
        "Customer ID (masked)" AS customer_id,
        "Offer ID" AS offer_id,
        COUNT(*) AS click_count
    FROM add_event
    GROUP BY customer_id, offer_id
""").df()


transactions = con.execute("""
    SELECT
        "Customer ID (masked)" AS customer_id,
        id8 AS offer_id,  -- assuming id8 = Offer ID in add_trans
        COUNT(*) AS trans_count
    FROM add_trans
    GROUP BY customer_id, offer_id
""").df()


# Outer merge to include all customers who either clicked or transacted
ctr_df = pd.merge(clicks, transactions, on=['customer_id', 'offer_id'], how='outer')



# Fill NaNs with 0
ctr_df.fillna(0, inplace=True)

# Calculate CTR (avoid division by zero)
ctr_df['CTR'] = ctr_df['click_count'] / (ctr_df['trans_count'] + ctr_df['click_count'])
ctr_df['CTR'] = ctr_df['CTR'].fillna(0).round(4)

# ----------------------
# 4️⃣ Final CTR Summary
# ----------------------
print("Click-Through Rate (CTR) per customer-offer pair:")
display(ctr_df.sort_values(by="CTR", ascending=False))

# Optional: Average CTR per customer
avg_ctr = ctr_df.groupby('customer_id')['CTR'].mean().reset_index().rename(columns={'CTR': 'avg_CTR'})
display(avg_ctr.sort_values(by="avg_CTR", ascending=False)) 
```

```python
# Register tables first (if using CSVs or loading)
# con.execute("CREATE VIEW train_data AS SELECT * FROM 'path_to_train.csv'")

con.execute("""
    CREATE OR REPLACE TABLE train_full AS
    SELECT 
        t.*,
        e."Event timestamp" AS event_timestamp,
        e.id6 AS event_type,
        m."Discount rate on the merchant offer" AS discount_rate,
        m.f377 AS offer_attr1,
        m.id10 AS offer_attr2,
        m.id11 AS offer_attr3,
        m.f378 AS offer_attr4,
        m.id12 AS offer_start,
        m.id13 AS offer_end,
        tr.f367 AS trans_amount,
        tr.f368 AS trans_type1,
        tr.f369 AS trans_type2,
        tr.f370 AS trans_date,
        tr.f371 AS trans_time,
        tr.f372 AS trans_loc,
        tr.id8 AS trans_offer_id
    FROM train_data t
    LEFT JOIN add_event e 
        ON t."Customer ID (masked)" = e."Customer ID (masked)" 
        AND t."Offer ID" = e."Offer ID"
    LEFT JOIN add_trans tr 
        ON t."Customer ID (masked)" = tr."Customer ID (masked)"
    LEFT JOIN offer_metadata m 
        ON t."Offer ID" = m."Offer ID"
""")
```

```python
df_model = con.execute("SELECT * FROM train_full").df()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

X = df_model.drop(['target_column', 'event_timestamp', 'Offer ID'], axis=1)  # Adjust columns
y = df_model['target_column']

X_train, X_val, y_train, y_val = train_test_split(X, y, stratify=y)

model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_val)
proba = model.predict_proba(X_val)[:,1]

print(classification_report(y_val, pred))
print("AUC-ROC:", roc_auc_score(y_val, proba))
```

```python
test_df = con.execute("SELECT * FROM test_full").df()
test_X = test_df[X.columns]  # Align features

preds = model.predict_proba(test_X)[:, 1]
submission = test_df[['Customer ID (masked)', 'Offer ID']].copy()
submission['prediction'] = preds
submission.to_csv("submission.csv", index=False)
```
